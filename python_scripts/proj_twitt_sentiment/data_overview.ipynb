```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df_twitter = pd.read_csv(
    "data/fourth_rendition_geolocated_id_posneutral_predict.csv",
    parse_dates=[
        'created_at'
    ]
).drop(columns={'id'})

df_twitter.created_at = df_twitter.created_at.dt.to_period(
    'Y'
)

df_twitter_count = df_twitter.groupby('created_at').count()
```

```python
df_built_wp = pd.read_csv('data/vindkraft_utbygd.csv')
df_built_wp['in_operation'] = pd.to_datetime(df_built_wp['forsteIdriftDato'],unit='ms').dt.to_period('Y')
df_built_wp = df_built_wp.groupby('in_operation').sum()
df_built_wp.assign(cumulative_wind = lambda x : x.effekt_MW_idrift.cumsum()).cumulative_wind
```

```python
df_twitter.loc[df_twitter['created_at'] == str(2022)].groupby('username').first().label.count()
```

```python
active_users = []

for year in df_twitter.created_at.unique():
    users_year = df_twitter.loc[df_twitter['created_at'] == str(year)].groupby('username').first().label.count()
    active_users.append(users_year)
```

```python
active_users
```

```python
table.assign(
    positive = df_twitter.groupby('created_at').sum().label,
    negative = lambda x : x.tweets - x.positive,
    share_negative = lambda x : round(x.negative / x.tweets,3)*100,
    new_users = df_twitter.groupby('username').first().reset_index().groupby('created_at').count().label,
    yearly_active_users = active_users,
    share_new_users = lambda x : round(x.new_users/x.yearly_active_users,2),
    tweets_per_user = lambda x : round(x.tweets/x.yearly_active_users,2),
)
```
