```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#import geopandas as gpd
import math
```

```python
df_twitter = pd.read_csv(
    "data/fourth_rendition_geolocated_id_posneutral_predict.csv",
    parse_dates=['created_at']
).sort_values(
    'created_at'
).groupby(
    'username'
).first(
).sort_values(
    'created_at'
).reset_index(
)

df_twitter['created_at'] = df_twitter['created_at'].dt.to_period('M')
```

```python

df_twitter
```

```python
fig, ax = plt.subplots(
    constrained_layout=False,
    #figsize=(10,8),
    dpi = 150
)

df_twitter.groupby('created_at').count().plot(
    ax = ax,
    y = 'label'
)

plt.ylabel('Number of new users')
plt.xlabel('Year')
plt.legend(['Monthly new users'])
```
